# KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4,n_init=20)
```

```python
km.fit(pc_data)
```

```python
cluster_labels=km.predict(pc_data)
```

```python
cluster_labels
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
plt.scatter(pc_data['pc1'], pc_data['pc2'], c = cluster_labels )
plt.xlabel("pc1")
plt.ylabel("pc2")
plt.title("pc1 Vs pc2")
plt.show()
```

```python
km.inertia_
```

```python
inertia_list = []


for i in range(1,11):
  km = KMeans(n_clusters = i, n_init = 20)
  km.fit(pc_data)
  inertia = km.inertia_
  inertia_list.append(inertia)
```

```python
inertia_list
```

```python
# Elbow method
```

```python
plt.plot(range(1,11), inertia_list)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show();
```

```python
# Hierarchial clustering
```

```python
#### scipy
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings = linkage(pc_data, method = "centroid")
```

```python
plt.figure(figsize= (10, 6))
dendrogram(mergings)
plt.title("Dendogram")
plt.axhline(92, color = "red", linestyle = "--")
plt.show()
```

```python
cluster_labels = fcluster(mergings, 92, criterion = "distance")
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
plt.scatter(pc_data['pc1'], pc_data['pc2'], c = cluster_labels )
plt.xlabel("pc1")
plt.ylabel("pc2")
plt.title("pc1 Vs pc2")
plt.show()
```
